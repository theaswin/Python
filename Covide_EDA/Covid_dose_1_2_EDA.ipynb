```python
fig,ax = plt.subplots(figsize=(50,20))
ax = sns.barplot(x = 'First_Dose',y = 'States',orient = 'h',data = df.sort_values('First_Dose',ascending=False))
```

```python
fig,ax = plt.subplots(1,sharey='row',figsize=(20,20))
sns.set_color_codes("pastel")
sns.barplot(x="First_Dose",y='States',data=df,label='Dose1',color='b')
sns.set_color_codes('muted')
sns.barplot(x='Second_Dose',y='States',data=df,label='Dose2',color='r')
```

```python
fig = plt.subplots(1,sharey="row",figsize=(55,6))

X = df.States
Y = df.First_Dose
Z = df.Second_Dose
ZZ = df["Second_Dose"] - df["First_Dose"]
df["net"] = df["First_Dose"] - df["Second_Dose"] 
YY = df.net

def subcategorybar(X, vals, width=0.8):
    n = len(vals)
    _X = np.arange(len(X))
    for i in range(n):
        plt.bar(_X - width/2. + i/float(n)*width, vals[i], 
                width=width/float(n), align="edge")   
    plt.xticks(_X, X)
    
subcategorybar(X, [Y,Z,YY])

plt.show()
```
